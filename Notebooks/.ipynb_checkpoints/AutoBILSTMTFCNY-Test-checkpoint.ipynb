```python
import os
import sys
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch.utils.data as Data
import numpy as np
import pandas as pd
import time
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
import seaborn as sns
from sparsemax import Sparsemax
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve, auc, recall_score
from sklearn.model_selection import train_test_split


# # Define functions to read TSV files
def read_tsv(filename, inf_ind, skip_1st=False, file_encoding="utf8"):
    extract_inf = []
    with open(filename, "r", encoding=file_encoding) as tsv_f:
        if skip_1st:
            tsv_f.readline()
        line = tsv_f.readline()
        while line:
            line_list = line.strip().split("\t")
            temp_inf = [line_list[ind] for ind in inf_ind]
            extract_inf.append(temp_inf)
            line = tsv_f.readline()
    return extract_inf

# Define a function that reads an amino acid feature file and generates a feature dictionary
def get_features(filename, f_num=15):
    f_list = read_tsv(filename, list(range(16)), True)
    f_dict = {}
    left_num = 0
    right_num = 0
    if f_num > 15:
        left_num = (f_num - 15) // 2
        right_num = f_num - 15 - left_num
    for f in f_list:
        f_dict[f[0]] = [0] * left_num + [float(x) for x in f[1:]] + [0] * right_num
    f_dict["X"] = [0] * f_num
    return f_dict

# Defining Input Functions
def generate_input(sps, sp_lbs, feature_dict, feature_num, ins_num, max_len):
    
    xs, ys, lens = [], [], []
    for i, sp in enumerate(sps):
        ys.append(sp_lbs[i])
        lens.extend([len(tcr[0]) if tcr[0] else 0 for tcr in sp])
        
    while len(lens) % ins_num != 0:
        lens = np.concatenate((lens, np.array([0])))  
    lens = np.array(lens)
    lens = lens.reshape(-1, ins_num)
    
    while lens.shape[0] < len(sps):
        lens = np.concatenate((lens, np.zeros((1, ins_num))), axis=0)
        
    for i, sp in enumerate(sps):
        x = [[[0] * feature_num for _ in range(max_len)] for _ in range(ins_num)]
        seq_count = 0  
        for j, tcr in enumerate(sp):
            tcr_seq = tcr[0]
            right_num = max_len - len(tcr_seq)
            tcr_seq += "X" * right_num
            tcr_matrix = []
            for aa in tcr_seq:
                tcr_matrix.append(feature_dict[aa.upper()])
            x[seq_count] = tcr_matrix
            seq_count += 1

        xs.append(x)

    xs = np.array(xs)
    xs = torch.tensor(xs, dtype=torch.float32)
    xs = xs.swapaxes(2, 3)
    ys = np.array(ys)
    ys = torch.tensor(ys, dtype=torch.float32).view(-1, 1)
    lens = torch.tensor(lens, dtype=torch.long)
    
    return xs, ys, lens

#Define the Generate Label function
def load_data(sample_dir):
    training_data = []
    training_labels = []
    for sample_file in os.listdir(sample_dir):
        training_data.append(read_tsv(os.path.join(sample_dir, sample_file), [0, 1], True))
        if "P" in sample_file:
            training_labels.append(1)
        elif "H" in sample_file:
            training_labels.append(0)
        else:
            print("Wrong sample filename! Please name positive samples with 'P' and negative samples with 'H'.")
            sys.exit(1)
        
    return training_data, training_labels

#Define the evaluation function
from tqdm import tqdm
def evaluate(model, criterion, test_loader, device='cuda'):
    test_total_loss = 0.0
    all_preds = []
    all_labels = []
    
    model.eval()
    with torch.no_grad():
        for test_batch_x, test_batch_y, test_valid_lens in test_loader:
            test_batch_x = test_batch_x.view(-1, 24, 15).to(device)
            test_batch_y = test_batch_y.to(device)
            test_pred = model(test_batch_x, test_valid_lens)

            test_loss = criterion(test_pred, test_batch_y)
            test_total_loss += test_loss.item()
            all_preds.append(test_pred.cpu().numpy())
            all_labels.append(test_batch_y.cpu().numpy())
            
        test_avg_loss = test_total_loss / len(test_loader)
        return test_avg_loss, all_preds, all_labels
    

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def metrics(all_preds, all_labels, threshold):
    # 将 logits 转换为概率
    all_probs = sigmoid(np.array(all_preds))
    # 生成二进制预测结果
    binary_preds = (all_probs > threshold).astype(int)
    conf_matrix = confusion_matrix(all_labels, binary_preds)
    accuracy = accuracy_score(all_labels, binary_preds)
    sensitivity = conf_matrix[1, 1] / (conf_matrix[1, 0] + conf_matrix[1, 1])
    specificity = conf_matrix[0, 0] / (conf_matrix[0, 0] + conf_matrix[0, 1])
    auc = roc_auc_score(all_labels, all_probs)
    
    return accuracy, sensitivity, specificity, auc


# Define CNNLayer
class CNNLayer(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, stride, padding,dropout_rate):
        super(CNNLayer, self).__init__()
        self.conv = nn.Conv1d(in_channels, out_channels, kernel_size, stride, padding)
        self.bn = nn.BatchNorm1d(out_channels)
        self.relu = nn.ReLU()

        self.dropout = nn.Dropout(dropout_rate)  
        self.adaptive_maxpool = nn.AdaptiveMaxPool1d(1)

    def forward(self, x):
        x = self.conv(x)
        x = self.bn(x)
        x = self.relu(x)
    
        x = self.dropout(x)  
        x = self.adaptive_maxpool(x)
        return x

#Define the dot product scoring function    
class DotProductScore(nn.Module):
    def __init__(self, hidden_size):
        super(DotProductScore, self).__init__()
        self.q = nn.Parameter(torch.empty(size=(hidden_size, 1), dtype=torch.float32))
        self.init_weights()
        
    def init_weights(self):
        
        initrange = 0.5
        self.q.data.uniform_(-initrange, initrange)

    def forward(self, inputs):

        scores = torch.matmul(inputs, self.q)
        scores = scores.squeeze(-1)
        
        return scores

# Define Attention  
class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.scores = DotProductScore(hidden_size)

    def forward(self, X, valid_lens):
        scores = self.scores(X)
        arrange = torch.arange(X.size(1), dtype=torch.float32, device=X.device).unsqueeze(0)
        mask = (arrange < valid_lens.unsqueeze(-1)).float()
        scores = scores * mask - (1 - mask) * 1e9
        attention_weights = nn.functional.softmax(scores, dim=-1)  
        out = torch.matmul(attention_weights.unsqueeze(1), X).squeeze(1)
        
        return out

# Define  AutoBILSTMTFCNY model   
class AutoBILSTMTFCNY(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers, dropout, ins_num):
        super(AutoBILSTMTFCNY, self).__init__()
        self.ins_num = ins_num
        
        self.cnn1 = CNNLayer(input_size, hidden_size * 2, kernel_size=8, stride=1, padding=3,dropout_rate=dropout)
        self.lstm = nn.LSTM(hidden_size * 2, hidden_size, num_layers, batch_first=True, dropout=dropout, bidirectional=True)
        self.attention = Attention(hidden_size * 2)
        self.cnn2 = CNNLayer(hidden_size * 2, hidden_size * 4, kernel_size=3, stride=1, padding=1,dropout_rate=dropout)  
        self.dropout = nn.Dropout(p=dropout)
        self.fc1 = nn.Linear(hidden_size * 4, output_size)
        
        
    def forward(self, seq, valid_lens):

        out = seq.permute(0, 2, 1)

        out = self.cnn1(out)

        out = out.permute(0, 2, 1) 

        output, _ = self.lstm(out)

        valid_lens = valid_lens.view(-1,).to(device)

        out = self.attention(output, valid_lens)

        out = out.unsqueeze(2)

        out = self.cnn2(out)
        
        out = out.view(out.size(0), -1)

        out = self.dropout(out)
        
        out = self.fc1(out)
        
        out = out.view(-1, self.ins_num, 1)  

        out = out.mean(dim=1) 

        return out

# Setting model parameters     
def init_model():
    input_size = 15  # Input feature dimensions
    hidden_size = 30  # Hide layer size
    num_layers = 3  # of layers in Bi-LSTM stacks
    output_size = 1  # Output size
    ins_num = 100  # 100 TCR sequences
    dropout = 0.6 # Dropout ratio
    
    return AutoBILSTMTFCNY(input_size, hidden_size, output_size, num_layers, dropout, ins_num)                
```

```python

# Introduce an early stop mechanism
sys.path.append('../')
from python_codes.pytorchtools import EarlyStopping


def process_disease(model_name, disease_name):
    # Reading model files
    model_path = f'../Model Test Files/AutoBILSTMTFCNY/{model_name}checkpoint.pt'
    # New Test Set Path
    new_data_dir = f'../New Test Data//{disease_name}/'
    # Read the amino acid characterization file
    aa_file = "../Data/PCA15.txt"
    aa_vectors = get_features(aa_file) 
    
    device="cuda"

    model = init_model().to(device)
    criterion = nn.BCEWithLogitsLoss()
    model.load_state_dict(torch.load(model_path))
    model.eval()

    new_data, new_labels = load_data(new_data_dir)
    new_input_batch, new_label_batch, new_valid_lens = generate_input(new_data, new_labels, aa_vectors, 15, 100, 24)
    
    new_input_batch, new_label_batch, new_valid_lens = new_input_batch.to("cuda"), new_label_batch.to("cuda"), new_valid_lens.to("cuda")
    new_dataset = TensorDataset(new_input_batch, new_label_batch, new_valid_lens)
    new_loader = DataLoader(new_dataset, batch_size=len(new_input_batch), shuffle=False)

    _, new_preds, new_labels = evaluate(model, criterion, new_loader, device=device)
    new_preds = np.concatenate(new_preds, axis=0)
    new_labels = np.concatenate(new_labels, axis=0)
    all_probs = sigmoid(np.array(new_preds))
    
    return new_labels,all_probs,new_preds

diseases = [ 'MS', 'T1D']
model_names = ["MS", 'T1D']
performance_results = []  

for disease, model_name in zip(diseases, model_names):
    new_labels, _,new_preds = process_disease(model_name, disease) 

    accuracy, sensitivity, specificity, auc_score = metrics(new_preds, new_labels, 0.5)

    performance_results.append({
        'Disease': disease,
        'Accuracy': accuracy,
        'Sensitivity': sensitivity,
        'Specificity': specificity,
        'AUC': auc_score
    })
    
 
    print(f"{disease} - Accuracy: {accuracy:.4f}, Sensitivity: {sensitivity:.4f}, Specificity: {specificity:.4f}, AUC: {auc_score:.4f}")


print("\nCollected Performance Metrics(threshold=0.5):")
for result in performance_results:
    print(result)
```

```python
# 绘制ROC曲线图
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
subplot_labels = ['a', 'b', 'c']  

for i, (disease, model_name) in enumerate(zip(diseases, model_names)):
    labels, probs,_ = process_disease(model_name, disease)
    fpr, tpr, _ = roc_curve(labels, probs)
    roc_auc = auc(fpr, tpr)
    axes[i].plot(1 - fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.4f}')
    axes[i].plot([1, 0], [0, 1], '--', color='grey')
    axes[i].set_xlim([1.0, 0.0])
    axes[i].set_ylim([0.0, 1.05])
    axes[i].set_xlabel('Specificity')
    axes[i].set_ylabel('Sensitivity')
    axes[i].set_title(f'{disease}')
    axes[i].legend(loc="lower right")
    axes[i].text(-0.1, 1.05, subplot_labels[i], transform=axes[i].transAxes, fontsize=16, fontweight='bold', va='top', ha='right')
    
plt.tight_layout()
# 保存图像
#save_path = ""
#plt.savefig(save_path, dpi=300)
plt.show()


```

```python

# 准备绘图，每种疾病一个图表
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
subplot_labels = ['a', 'b', 'c']  # 标签列表

for i, (disease, model_name) in enumerate(zip(diseases, model_names)):
    labels, _ ,probs = process_disease(model_name, disease)
    
    # 计算性能指标并寻找最佳阈值
    thresholds = np.arange(0, 1, 0.1)
    acc_scores = []
    sens_scores = []
    spec_scores = []
    auc_scores = []
    combined_scores = []

    for thresh in thresholds:
        accuracy, sensitivity, specificity, auc_score = metrics(probs, labels, thresh)
        acc_scores.append(accuracy)
        sens_scores.append(sensitivity)
        spec_scores.append(specificity)
        auc_scores.append(auc_score)
        combined_scores.append((accuracy + sensitivity + specificity) / 3)
        

    # 寻找综合得分最高的阈值
    best_index = np.argmax(combined_scores)
    best_threshold = thresholds[best_index]
    best_accuracy = acc_scores[best_index]
    best_sensitivity = sens_scores[best_index]
    best_specificity = spec_scores[best_index]
    best_auc = auc_scores[best_index]

    # 寻找具有相同最高综合得分的所有阈值
    max_combined_score = max(combined_scores)
    best_thresholds = [thresh for i, thresh in enumerate(thresholds) if combined_scores[i] == max_combined_score]

    # 获取这些阈值的范围
    best_threshold_range = (min(best_thresholds), max(best_thresholds))

    # 绘制阈值与性能指标的曲线
    ax = axes[i]
    ax.plot(thresholds, acc_scores, 'b', label='Accuracy')
    ax.plot(thresholds, sens_scores, 'g', label='Sensitivity')
    ax.plot(thresholds, spec_scores, 'r', label='Specificity')
    ax.axvline(x=best_threshold_range[0], color='purple', linestyle='--', label=f'Best Threshold Range = {best_threshold_range[0]:.2f} to {best_threshold_range[1]:.2f}')
    ax.axvline(x=best_threshold_range[1], color='purple', linestyle='--')
    ax.scatter(best_threshold, best_accuracy, color='blue', s=50, label=f'Best Accuracy = {best_accuracy:.4f}')
    ax.scatter(best_threshold, best_sensitivity, color='green', s=50, label=f'Best Sensitivity = {best_sensitivity:.4f}')
    ax.scatter(best_threshold, best_specificity, color='red', s=50, label=f'Best Specificity = {best_specificity:.4f}')
    ax.scatter(best_threshold, best_auc, color='darkorange', s=50, label=f'Best AUC = {best_auc:.4f}')
    ax.fill_betweenx([0, 1], best_threshold_range[0], best_threshold_range[1], color='purple', alpha=0.3)
    ax.set_xlabel('Threshold')
    ax.set_ylabel('Score')
    ax.set_title(f'{disease} ')
    ax.legend(loc="lower right")
    ax.grid(True)
      # 将标签放置在图外左上角
    axes[i].text(-0.1, 1.05, subplot_labels[i], transform=axes[i].transAxes, fontsize=16, fontweight='bold', va='top', ha='right')

plt.tight_layout()
save_path = '../results/AttenBILSTMCNY-combined_figure_Thresholds.png'  # 确保目录存在或根据需要修改
plt.savefig(save_path, dpi=300)
plt.show()
```
